```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_requests(url):
    try:
        r = requests.get(url)
        print(r.text)
    except:
        print("Can't get the request")

def scraping(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
        
        with open("scraped_data.txt", "w") as f:
            for i in soup.find_all(class_=["title", "description card-text", "ratings", "caption"]):
                f.write(str(i))
                
        print("Scraped")
        
    except (FileNotFoundError, NameError, SyntaxError) as e:
        print(e)

url = "https://webscraper.io/test-sites/e-commerce/allinone"

get_requests(url)
scraping(url)
```
